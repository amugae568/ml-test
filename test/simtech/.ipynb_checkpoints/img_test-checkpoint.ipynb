```python
train_path = 'train'
labels = os.listdir(train_path)
X = []
y = []
for label in labels:
    path = os.path.join(train_path, label)
    image_names = os.listdir(path)

    for image_name in image_names:
        img = cv2.imread(os.path.join(path, image_name))
        img.resize(224, 224, 3)
        #gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        #processed = cv2.normalize(gray_img, None, 0, 255, cv2.NORM_MINMAX)
        #img = np.concatenate([img, processed[:, :, np.newaxis]], axis=2)
        X.append(img)
        y.append(label)
y = list(map(lambda x: int(x) - 1, y))
```

```python
X = np.array(X)/255
y = np.array(y)
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,stratify=y,random_state=42, test_size=0.2)
# X_train, y_train, X_valid, y_valid = train_test_split(X_train, y_train,stratify=y,random_state=42, test_size=0.2)
```

```python
y_train, y_test = tf.one_hot(y_train, 3), tf.one_hot(y_test, 3)
```

```python
from tensorflow.keras.applications.efficientnet_v2 import EfficientNetV2B0
```

```python
efficientNet = EfficientNetV2B0()
```

```python
for layer in efficientNet.layers:
    if layer.name.startswith('block3'):
        break
    layer.trainable = False
    
```

```python
efficientNetOutput = efficientNet.layers[-1].output
x = tf.keras.layers.Flatten()(efficientNetOutput)
x = tf.keras.layers.Dense(512, activation='relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
x = tf.keras.layers.Dense(3, activation='softmax')(x)
efficientTest = tf.keras.models.Model(inputs=efficientNet.input, outputs=x)
efficientTest.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
efficientTest.summary()
```

```python
history = efficientTest.fit(X_train, y_train, epochs=10, batch_size=16,
                    validation_data=(X_test, y_test))
```
